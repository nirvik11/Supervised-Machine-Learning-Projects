```python
import numpy as np
import pandas as pd
```

```python
matches = pd.read_csv('/content/matches.csv')
delivery = pd.read_csv('/content/deliveries.csv')
```

```python
matches.head()
```

```python
matches.shape
```

```python
delivery.head()
```

```python
total_score = delivery.groupby(['match_id', 'inning']).sum()['total_runs'].reset_index()
```

```python
total_score
```

```python
total_score = total_score[total_score['inning'] == 1]
```

```python
total_score
```

```python
match_df = matches.merge(total_score[['match_id','total_runs']], left_on = 'id', right_on = 'match_id')
```

```python
match_df
```

```python
match_df['team1'].unique()
```

```python
teams = [
    'Sunrisers Hyderabad',
    'Mumbai Indians',
    'Royal Challengers Bangalore',
    'Kolkata Knight Riders',
    'Chennai Super Kings',
    'Delhi Capitals',
    'Rajasthan Royals',
    'Kings XI Punjab'
]
```

```python
match_df['team1'] = match_df['team1'].str.replace('Delhi Daredevils', 'Delhi Capitals')
match_df['team2'] = match_df['team2'].str.replace('Delhi Daredevils', 'Delhi Capitals')

match_df['team1'] = match_df['team1'].str.replace('Deccan Chargers', 'Sunrisers Hyderabad')
match_df['team2'] = match_df['team2'].str.replace('Deccan Chargers', 'Sunrisers Hyderabad')
```

```python
match_df = match_df[match_df['team1'].isin(teams)]
match_df = match_df[match_df['team2'].isin(teams)]
```

```python
match_df.shape
```

```python
match_df = match_df[match_df['dl_applied'] == 0]
```

```python
match_df
```

```python
match_df = match_df[['match_id', 'city', 'winner', 'total_runs']]
```

```python
delivery_df = match_df.merge(delivery, on = 'match_id')
```

```python
delivery_df
```

```python
delivery_df = delivery_df[delivery_df['inning'] == 2]
```

```python
delivery_df.shape
```

```python
delivery_df['current_score'] = delivery_df.groupby('match_id').cumsum()['total_runs_y']
```

```python
delivery_df
```

```python
delivery_df['runs_left'] = delivery_df['total_runs_x'] - delivery_df['current_score']
```

```python
delivery_df['balls_left'] = 126 - (delivery_df['over']*6 + delivery_df['ball'])
```

```python
update_func = lambda x: 0 if pd.isna(x) else 1
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(update_func)
wickets = delivery_df.groupby('match_id').cumsum()['player_dismissed'].values
delivery_df['wickets'] = 10 - wickets
delivery_df.head()
```

```python
delivery_df.tail()
```

```python
delivery_df['crr'] = (delivery_df['current_score']*6)/(120 - delivery_df['balls_left'])
```

```python
delivery_df['rrr'] = (delivery_df['runs_left']*6)/delivery_df['balls_left']
```

```python
delivery_df
```

```python
def result(row):
  return 1 if row['batting_team'] == row['winner'] else 0

delivery_df['result'] = delivery_df.apply(result, axis = 1)
```

```python
delivery_df
```

```python
final_df = delivery_df[['batting_team', 'bowling_team', 'city', 'runs_left', 'balls_left', 'wickets', 'crr', 'rrr', 'result']]
```

```python
final_df = final_df.sample(final_df.shape[0])
```

```python
final_df.dropna(inplace = True)
```

```python
final_df = final_df[final_df['balls_left'] != 0]
```

```python
final_df.sample()
```

```python
X = final_df.drop('result', axis=1)
Y = final_df['result']
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 1)
```

# ONE HOT ENCODING

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer([('trf', OneHotEncoder(sparse = False, drop = 'first'), ['batting_team','bowling_team','city'])], remainder = 'passthrough') 
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
```

```python
model = Pipeline(steps = [
    ('step1', trf),
    ('step2', LogisticRegression())
])
```

```python
model.fit(X_train, Y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
Y_pred = model.predict(X_test)
accuracy_score(Y_pred, Y_test)
```

```python
import pickle
pickle.dump(model, open('ipl.pkl', 'wb'))
```
